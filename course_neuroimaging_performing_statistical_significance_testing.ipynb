## Linear Regression t-Test

We consider a dataset consisting of 20 rows, each representing height and weight related to a particular student.

We perform a regression t-test to examine the relationship between two variables: the `height` as the predictor variable and the `weight` as the response variable.

We consider a linear regression model described by the following equation:

$$y=\beta_0 + \beta_1x + \epsilon$$

**State hypothesis:**
* Null hypothesis: $\beta_1=0$.
* Alternative hypothesis: $\beta_1 \neq 0$.

We perform a regression t-test using the **`linregress()`** function provided by the **`stats`** module from the **`scipy`** library.

The the **`linregress()`** function calculates a linear regression for two sets of measurements, an independent variable **`x`** and a dependent variable **`y`**.

It accepts as arguments two arrays of the same length, **`x`** and **`y`**.

It returns a **`LinregressResult`** object containing the following attributes:
* **`slope`**: The slope of the regression line.
* **`intercept`**: The y-intercept of the regression line.
* **`rvalue`**: The correlation coefficient ($R$-value). The square of the $R$-value is the coefficient of determination ($R^2$).
* **`pvalue`**: The two-sided p-value for a hypothesis test whose null hypothesis is that the slope is zero.
* **`stderr`**: The standard error of the estimated slope.
* **`intercept_stderr`**: The standard error of the estimated intercept (available in newer SciPy versions).

## Imports

```python
import pandas as pd
import seaborn as sns
from scipy.stats import linregress
```

### Load Dataset

We use a dataset consisting of 20 rows, each representing height and weight related to a particular student.

```python
# dataset url
url = 'https://raw.githubusercontent.com/tariqzahratahdi/DataScience/refs/heads/main/datasets/students_height_weight.csv'

# load dataset
data = pd.read_csv(url)

# print dataset head
data.head()
```

We plot the data as a scatter plot to visualize the underlying linear relationship between the two variables.

```python
# scatter plot
sns.scatterplot(data = data, x = "height", y = "weight")
```

### Perform t-Test

We perform a linear regression t-test using the **`linregress()`** function provided by the **`stats`** module from the **`scipy`** library.

```python
# perform linear regression t-test
test_result = linregress(data["height"], data["weight"])

# print results
print(f'p-value: {test_result.pvalue:.2e}')
print(f'slope: {test_result.slope:.2f}')
print(f'intercept: {test_result.intercept:.2f}')
```

Interpret the result of the t-test:

We reject the null hypothesis since the p-value is smaller than the specified significance level `alpha` set to 0.05.

```python
# set significance level
alpha = 0.05

# test null hypothesis
if (test_result.pvalue < alpha):
  print('reject the null hypothesis.')
else:
  print('fail to reject the null hypothesis.')
```
